```python
A = np.array([[4, 1],
              [2, 3]], dtype=float)

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify A v = λv
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    left = A @ v
    right = lam * v

    print(f"\nEigenpair {i+1}")
    print("A @ v =", left)
    print("λ * v =", right)

    assert np.allclose(left, right)

# Plot
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i] * eigenvalues[i]

    ax.quiver(
        0, 0,
        v[0], v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f"λ={eigenvalues[i]:.2f}"
    )

ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

ax.axhline(0,color='k')
ax.axvline(0,color='k')

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Eigenvectors scaled by Eigenvalues")

ax.legend()

plt.grid()
plt.show()

print("4.3 passed ✓")
```

```python
y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

# Errors
errors = y_true - y_pred

# MAE
mae = np.mean(np.abs(errors))

# MSE
mse = np.mean(errors ** 2)

# RMSE
rmse = np.sqrt(mse)

# R²
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

r2 = 1 - (ss_res / ss_tot)

# Adjusted R²
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
```

```python
from statsmodels.tsa.stattools import adfuller
np.random.seed(0)

s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test against standard normal
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(
    f"KS test s1 (normal):      "
    f"stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}"
)

print(
    f"KS test s2 (exponential): "
    f"stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}"
)

# Non-stationary time series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

# ADF test
adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once
ts_diff = np.diff(ts)

# ADF again
adf_diff = adfuller(ts_diff)

print(
    f"\nAfter differencing — "
    f"p-value: {adf_diff[1]:.4f}, "
    f"Stationary? {adf_diff[1] < 0.05}"
)
```

```python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) *
              ln(actual_pct / expected_pct))
    """

    epsilon = 1e-10

    # Create bins from expected distribution
    breakpoints = np.histogram_bin_edges(expected, bins=bins)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    expected_pct += epsilon
    actual_pct += epsilon

    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi


np.random.seed(1)

train_dist = np.random.normal(50, 10, 1000)
drift_dist = np.random.normal(65, 12, 1000)

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")

print(
    f"Shift severity: "
    f"{'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}"
)

# Plot
fig, ax = plt.subplots()

ax.hist(
    train_dist,
    bins=30,
    alpha=0.6,
    density=True,
    label="Training Distribution"
)

ax.hist(
    drift_dist,
    bins=30,
    alpha=0.6,
    density=True,
    label="Production Distribution"
)

ax.legend()
ax.set_title("Distribution Shift Visualization")

plt.tight_layout()
plt.show()
```

```python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (
    p_free_given_spam * p_spam
    + p_free_given_ham * p_ham
)

# Bayes
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam,
                        p_word_given_spam,
                        p_word_given_ham):

    prior_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam * prior_spam
        + p_word_given_ham * prior_ham
    )

    posterior = (
        p_word_given_spam * prior_spam
    ) / evidence

    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
```

```python
np.random.seed(7)

population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# Draw 5000 samples of size 30 and compute their means
sample_means = np.array([
    np.random.choice(population, size=sample_size, replace=True).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std = population.std()

# CLT predicted standard deviation
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots(figsize=(8, 5))

# Histogram of sample means
ax.hist(sample_means,
        bins=40,
        density=True,
        alpha=0.7,
        label='Sample Means')

# Theoretical Normal PDF from CLT
x = np.linspace(sample_means.min(),
                sample_means.max(),
                500)

ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.set_title('CLT: Distribution of Sample Means')
ax.legend()

plt.tight_layout()
plt.show()

# KS Test
ks_result = stats.kstest(
    sample_means,
    'norm',
    args=(pop_mean, clt_std)
)

print(
    f"KS test p-value: {ks_result.pvalue:.4f} "
    f"→ Approximately normal? {ks_result.pvalue > 0.05}"
)
```

```python
#The Central Limit Theorem (CLT) is important because it tells us that the distribution of sample means becomes approximately normal as the sample size grows, even when the original data is not normally distributed. This allows statisticians and machine learning practitioners to make reliable inferences about populations using sample data.

#The CLT justifies the normality assumption of sampling distributions used in many classical statistical methods, such as confidence intervals, hypothesis tests (z-tests and many t-test applications), and regression inference.
```
